```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('Amazon sales data.csv')
```

```python
df.head(20)
```

```python
df['Order Date'] = df['Order Date'].str.replace('-', '/')
df['Ship Date'] = df['Ship Date'].str.replace('-', '/')
```

```python
df.head(20)
df = df.astype(str)
columns = df.columns.tolist()
print(columns)
for column in columns:
    df[column] = df[column].str.strip()
```

```python
df.head(20)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
df['Order Date'] = df['Order Date'].astype(str)
df['Year'] = df['Order Date'].apply(lambda x: x.split('-')[0])
df['Month'] = df['Order Date'].apply(lambda x: x.split('-')[1])
```

```python
df.head()
```

```python
for column in columns[8:]:
    df[column] = df[column].astype(float)
```

```python
df.head()
```

```python
df = df.drop(['Order ID'],axis=1)
```

```python
df.head()
```

```python
df = df.drop(['Order Date'],axis=1)
```
